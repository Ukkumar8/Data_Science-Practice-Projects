**Data Cleaning**

Load the data set into Data frame

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
cdc=pd.read_csv('Cardiotocographic.csv')
cdc.head()
```

Identifing the missing values and checking null values in the given data set

```python
cdc.info()
```

Removing the duplicate values in the given data set

```python
cdc.duplicated().sum()
```

```python
cdc.drop_duplicates(inplace=True)
```

Finding the null values in data set

```python
cdc.isnull().sum()
```

We have find outliers using Box plot

```python
for i in cdc.columns:
  sns.boxplot(cdc[i])
  plt.show()
```

We have filled null values using median

```python
columns=['LB','AC','FM','UC','DL','DS','DP','ASTV','MSTV','ALTV','MLTV','Width','Tendency','NSP']
for col in columns:
  cdc[col]=cdc[col].fillna(cdc[col].median())
```

```python
cdc.isnull().sum()
```

Histogram has ploted to find weather it is positively skewed or negatively skewed

```python
cdc.hist(figsize=(20,20))
plt.show()
```

Here the correlation  

```python
plt.figure(figsize=(12,10))
sns.heatmap(cdc.corr(),annot=True)
```
